## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
application_df = pd.read_csv("https://static.bc-edx.com/data/dla-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(['EIN', 'NAME'], axis=1)
```

```python
# Determine the number of unique values in each column.
unique_counts = application_df.nunique()
print(unique_counts)
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()
print(application_type_counts)
```

```python
# Choose a cutoff value and create a list of application types to be replaced
cutoff_value = 500

# use the variable name `application_types_to_replace`
application_types_to_replace = application_df['APPLICATION_TYPE'].value_counts()[application_df['APPLICATION_TYPE'].value_counts() < cutoff_value].index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_counts = application_df['CLASSIFICATION'].value_counts()
print(classification_counts)
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_counts_gt1 = classification_counts[classification_counts > 1]
print(classification_counts_gt1)
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
cutoff_value = 500

# Use the variable name `classification_to_replace`
classification_to_replace = classification_counts[classification_counts < cutoff_value].index.tolist()

# Replace in the dataframe
for cls in classification_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls, "Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df = pd.get_dummies(application_df)

# Display the first few rows of the dataframe
application_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df['IS_SUCCESSFUL']
X = application_df.drop(columns=['IS_SUCCESSFUL'])

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Original Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_of_input_features = X_train_scaled.shape[1]
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=80, activation='relu', input_dim=number_of_input_features))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=30, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
history = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

## Build the Optimization Neural Network Models

### Adjust the input data to ensure that no variables or outliers are causing confusion in the model

```python
# Split data into features and target
y = application_df['IS_SUCCESSFUL']
X = application_df.drop(columns=['IS_SUCCESSFUL'])
```

```python
# Split data into training and testing datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, stratify=y)
```

```python
# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

#### Make three attempts at optimising the model. 

### Optimization Attempt 1

```python
# Define the model
model_attempt_1 = tf.keras.models.Sequential([
    tf.keras.layers.Dense(units=100, activation='relu', input_dim=X_train_scaled.shape[1]),
    tf.keras.layers.Dense(units=50, activation='relu'),
    tf.keras.layers.Dense(units=25, activation='relu'),
    tf.keras.layers.Dense(units=1, activation='sigmoid')
])

# Compile the model
model_attempt_1.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Implement early stopping
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)

# Train the model
history_1 = model_attempt_1.fit(X_train_scaled, y_train, epochs=150, validation_data=(X_test_scaled, y_test), callbacks=[early_stopping])

# Evaluate the model
model_loss_1, model_accuracy_1 = model_attempt_1.evaluate(X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss_1}, Accuracy: {model_accuracy_1}")
```

### Optimization Attempt 2

```python
from tensorflow.keras.optimizers import RMSprop

# Adjust learning rate
learning_rate = 0.001

# Build model 2
model_attempt_2 = tf.keras.models.Sequential([
    tf.keras.layers.Dense(units=150, activation='tanh', input_dim=X_train_scaled.shape[1]),
    tf.keras.layers.Dense(units=100, activation='relu'),
    tf.keras.layers.Dense(units=50, activation='tanh'),
    tf.keras.layers.Dense(units=1, activation='sigmoid')
])

# Compile the model with a different optimizer
model_attempt_2.compile(loss='binary_crossentropy', optimizer=RMSprop(learning_rate=learning_rate), metrics=['accuracy'])

# Train the model
history_2 = model_attempt_2.fit(X_train_scaled, y_train, epochs=150, validation_data=(X_test_scaled, y_test), callbacks=[early_stopping])

# Evaluate the model
model_loss_2, model_accuracy_2 = model_attempt_2.evaluate(X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss_2}, Accuracy: {model_accuracy_2}")
```

### Optimization Attempt 3

```python
# Build model 3
model_attempt_3 = tf.keras.models.Sequential([
    tf.keras.layers.Dense(units=200, activation='relu', input_dim=X_train_scaled.shape[1]),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(units=100, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(units=50, activation='relu'),
    tf.keras.layers.Dense(units=1, activation='sigmoid')
])

# Compile the model
model_attempt_3.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
history_3 = model_attempt_3.fit(X_train_scaled, y_train, epochs=100, batch_size=20, validation_data=(X_test_scaled, y_test), callbacks=[early_stopping])

# Evaluate the model
model_loss_3, model_accuracy_3 = model_attempt_3.evaluate(X_test_scaled, y_test, verbose=2)
print(f"Loss: {model_loss_3}, Accuracy: {model_accuracy_3}")
```

## BEST MODEL

```python
# Choose the best model
best_model = max([(model_attempt_1, model_accuracy_1), (model_attempt_2, model_accuracy_2), (model_attempt_3, model_accuracy_3)], key=lambda x: x[1])[0]

# Save the best model
best_model.save("AlphabetSoupCharity_Optimisation.h5")
```
